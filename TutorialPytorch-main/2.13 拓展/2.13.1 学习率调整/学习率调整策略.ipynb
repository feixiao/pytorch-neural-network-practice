```python
import torch
import torch.optim as optim
import torch.nn as nn
import matplotlib.pyplot as plt

class Net(nn.Module):
    def __init__(self):
        super().__init__()
        self.linear = nn.Linear(1, 1)
        
    def forward(self, x):
        return sele.linear(x)
net = Net()


# 1：定义优化器
optimizer_ExpLR_098 = torch.optim.SGD(net.parameters(), lr=0.1)
optimizer_ExpLR_090 = torch.optim.SGD(net.parameters(), lr=0.1)
optimizer_ExpLR_099 = torch.optim.SGD(net.parameters(), lr=0.1)
# 2：给优化器绑定学习率衰减策略
ExpLR_098 = torch.optim.lr_scheduler.ExponentialLR(optimizer_ExpLR_098, gamma=0.98)
ExpLR_090 = torch.optim.lr_scheduler.ExponentialLR(optimizer_ExpLR_090, gamma=0.9)
ExpLR_099 = torch.optim.lr_scheduler.ExponentialLR(optimizer_ExpLR_099, gamma=0.99)
# 3：定义步长和对应学习率值的容器
step_list = []
ExpLR_098_list = []
ExpLR_090_list = []
ExpLR_099_list = []
num_steps = 1000 # 总迭代次数
for i in range(num_steps):
    # 4：更新学习率
    ExpLR_098.step()
    ExpLR_090.step()
    ExpLR_099.step()
    # 5：记录当前步数对应的学习率
    step_list.append(i)
    ExpLR_098_list.append(optimizer_ExpLR_098.param_groups[0]["lr"])
    ExpLR_090_list.append(optimizer_ExpLR_090.param_groups[0]["lr"])
    ExpLR_099_list.append(optimizer_ExpLR_099.param_groups[0]["lr"])

plt.figure(figsize=(4, 3))
plt.plot(step_list, ExpLR_098_list, label="gamma=0.98", color="red")
plt.plot(step_list, ExpLR_090_list, label="gamma=0.90")
plt.plot(step_list, ExpLR_099_list, label="gamma=0.99")
plt.legend(loc="upper right")
plt.show()
```
