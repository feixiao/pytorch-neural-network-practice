```python
import torchvision.transforms as tt
from PIL import Image
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
# 用于显示中文字体
mpl.rcParams['font.sans-serif']=['SimHei']
mpl.rcParams['axes.unicode_minus']=False
# 放在data文件夹下的演示图像
image = Image.open("data/tree.jpg")
```

```python
def random_crop(img, crop_size):
    # crop_size代表裁剪后的图像尺寸
    RC = tt.RandomCrop(crop_size)
    rc_img = RC(img)
    return rc_img 

def horizontal_flip(img):
    HF = tt.RandomHorizontalFlip()
    hf_img = HF(img)
    return hf_img

def random_rotation(img, degrees):
    # degrees代表随机旋转的角度范围
    RR = tt.RandomRotation(degrees)
    rr_img = RR(img)
    return rr_img
```

```python
rc_img = random_crop(image, (150, 150))
hf_img = horizontal_flip(image)
rr_img = random_rotation(image, (10, 80))
image_list = [image, rc_img, hf_img, rr_img]
title_list = ["原始图片", "随机裁剪:rc_img", "水平翻转:hf_img", "随机旋转:rr_img"]

plt.figure(figsize=(10, 10))
for i in range(4): 
    plt.subplot(2, 2, i+1)
    # 设置坐标轴字体刻度
    plt.xticks(fontsize=20)
    plt.yticks(fontsize=20)
    plt.title(title_list[i], fontsize=25)
    plt.imshow(np.array(image_list[i]))
plt.savefig("random.png", dpi=500, bbox_inches="tight")
plt.show()
```
