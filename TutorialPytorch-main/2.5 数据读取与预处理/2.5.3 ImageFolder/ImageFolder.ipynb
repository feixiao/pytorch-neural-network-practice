```python
from torchvision import transforms
from torchvision.datasets import ImageFolder
import matplotlib.pyplot as plt

image_folder = ImageFolder(root="data_class/")
print("all classes: ", image_folder.classes)
print("class with label index: ", image_folder.class_to_idx)
print("path of images: ", image_folder.imgs)
# 输出结果：
# ['men', 'women']
# {'men': 0, 'women': 1}
# [('data_class/men\\1.jpg', 0), ('data_class/women\\1.jpg', 1)]
```

```python
for data in image_folder:
    print(data)
# 输出结果：
# (<PIL.Image.Image image mode=RGB size=250x250 at 0x1F882513198>, 0)
# (<PIL.Image.Image image mode=RGB size=600x800 at 0x1F882513358>, 1)
```

```python
# 分别获取图像和标签
img_0, class_0 = image_folder[0]
img_1, class_1 = image_folder[1]

# 展示图像和对应的标签
# plt.figure(figsize=(7, 7))
plt.subplot(1, 2, 1)
plt.imshow(img_0)
plt.title(image_folder.classes[0])
plt.subplot(1, 2, 2)
plt.imshow(img_1)
plt.title(image_folder.classes[1])

plt.savefig("men_women.png", dpi=500, bbox_inchs="tight")
plt.show()
```
