```python
import PIL
from PIL import Image

# 读取本地图片
image = Image.open("desk.jpg")
print("image width ",image.width)
print("image height", image.height)

image = image.resize(size=(256, 256), resample=PIL.Image.NEAREST)
print("image width ",image.width)
print("image height", image.height)

# 输出结果：
# image width  400
# image height 300
# image width  256
# image height 256
```

```python
import numpy as np

# 创建数值为[1, 2, 3,.., 255]的数组
line = np.linspace(0, 255, 256)
line = np.expand_dims(line, 0)
image = np.tile(line, (256, 1))
print("Image shape ", image.shape)
# 输出结果
# Image shape  (256, 256)

image = image.astype(np.uint8)
# 从numpy创建Image对象
image = Image.fromarray(image)
# 保存图片
image.save("gradual_light.jpg")
```

```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

image = Image.open("desk.jpg")
# 将图片转为数组类型
image = np.array(image)
plt.figure(figsize=(6, 5))

# 子图的行列数量
rows = 2
cols = 2

for i in range(rows * cols):
    # 创建子图
    plt.subplot(rows, cols, i + 1)
    # 设置子图的标题和内容
    plt.title("Figure " + str(i))
    plt.imshow(image)

# 指定保存的名称和分辨率
plt.savefig("9_desks.jpg", dpi=500, bbox_inches='tight')
plt.show()
```
