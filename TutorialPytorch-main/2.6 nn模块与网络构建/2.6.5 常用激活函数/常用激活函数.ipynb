```python
import torch.nn as nn
import torch

# 定义激活函数
relu = nn.ReLU()
x = torch.linspace(-1, 1, 5)
# 计算激活输出
out = relu(x)
print(x, "\n", out)
# tensor([-1.0000, -0.5000,  0.0000,  0.5000,  1.0000]) 
# tensor([0.0000, 0.0000, 0.0000, 0.5000, 1.0000])
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 定义十二种激活函数
acti = [nn.ReLU(), nn.LeakyReLU(0.1), nn.ELU(alpha=1), nn.PReLU(), nn.ReLU6(),
        nn.RReLU(), nn.SELU(), nn.CELU(), nn.Sigmoid(),nn.Tanh(),
        nn.LogSigmoid(), nn.Tanhshrink()]

# 记录每种激活函数对应的名字
titles = ["ReLU", "LeakyReLU", "ELU", "PReLU", "ReLU6",
        "RReLU", "SELU", "CELU", "Sigmoid", "Tanh",
        "LogSigmoid", "Tanhshrink"]

# 子图布局为4行3列
rows = 4
cols = 3
plt.figure(figsize=(8, 8))

for i in range(rows * cols):
    # 创建输入Tensor的数值范围是-10到10，等间隔取1000个点
    x = torch.linspace(-10, 10, 1000)
    # 计算激活后的输出
    out = acti[i](x)
    # 将输入和输出Tensor转为numpy
    x = x.detach().numpy()
    y = out.detach().numpy()
    # 创建子图，并绘制XY曲线
    plt.subplot(rows, cols, i+1)
    plt.tight_layout(pad=0.2) 
    plt.title(titles[i])
    plt.plot(x, y)
plt.savefig("acti.png")
plt.show()
```
